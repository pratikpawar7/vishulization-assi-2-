```python
'''
Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
'''


Seaborn is a popular data visualization library in Python that provides a high-level interface
for creating informative and attractive statistical graphics. 
Here are five types of plots that we can create using Seaborn, along with their uses:

Scatter plots - Scatter plots are used to visualize the relationship between two continuous variables. 
                They can help identify patterns, trends, or outliers in the data.

Line plots - Line plots are used to show the trend or change of one or more variables over time or other continuous variables.

Bar plots - Bar plots are used to compare the values of different categories of data. 
            They are particularly useful for showing changes over time or for comparing data across different groups.

Histograms - Histograms are used to display the distribution of a continuous variable. 
            They can help identify the shape, center, and spread of the data.

Heat maps - Heat maps are used to display a matrix of values with colors that represent the magnitude of each value.
            They are particularly useful for visualizing patterns or correlations in large datasets.
```

```python
'''
Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.
'''

import  seaborn as sns 

fmri = sns.load_dataset('fmri')
print(fmri)
sns.scatterplot(x  = fmri.timepoint , y =fmri.signal ) 
```

```python
'''
Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.
'''

titanic = sns.load_dataset('titanic')
print(titanic)
t1  =sns.scatterplot(x = titanic.pclass, y = titanic.age) 

t1
```

```python
'''
Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.
'''
import seaborn as sns 
diamond = sns.load_dataset('diamonds')
d1 = sns.scatterplot(x = diamond.price , y = diamond.cut ,data = diamond ,  hue = 'cut') 
d1 
```

```python
'''
Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.
'''

import seaborn as sns 
iris = sns.load_dataset('iris')
i1 = sns.pairplot(iris , hue = 'species')
```
